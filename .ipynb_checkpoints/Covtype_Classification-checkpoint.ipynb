## Data Preprocessing

```python
# Method in Caruana Paper: Largest class(7) is positive and everything else is negative
def transform_type(covtype):
    if covtype == 7:
        return 1
    else:
        return 0
```

```python
data['cover_type'] = data['cover_type'].apply(transform_type)
```
